```python
#standard scaling
```

```python
df= data.drop(['country'],axis=1)
cols= df.columns
from sklearn.preprocessing import StandardScaler
st_scale= StandardScaler()
df=st_scale.fit_transform(df)
```

```python
x=pd.DataFrame(df,columns=[cols])
x
```

```python
#kmean clustering
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss=[]
silhouette_coefficients = []
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=32)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    score = silhouette_score(x, kmeans.labels_)
    silhouette_coefficients.append(score)
    print('WCSS-Cluster',i,'=',kmeans.inertia_)
```

```python
#Elbow method
```

```python
plt.plot(range(2,11),wcss)
plt.title('Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS/SSE')
plt.show()
```

```python
#no of clusters using silhouette_coefficient
```

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
#fitting kmeans with no of clusters as 4
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=32)
kmeans.fit(x)
cluster=kmeans.labels_
cluster
```
